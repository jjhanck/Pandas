**Overview**:  Final dataset as analyzed had 16 attributes.  Above are the grouped results of this analysis.  The grouping variables are QIs, cluster_method, and clusters.  
1. QIs is the number of columns selected for clustering (2-16).  
2. Cluster_method is the binning strategy (fixed-member = 0, fixed-width = 1).  
3. Clusters is the level of generalization/coarsening provided by the cluster method (2,10,30,100,500).  For instance, a value of 100 means that each value in a selected columns is assigned to 1 of 100 discrete values.

All together now.  A row in the above table with say (QI = 15, Cluster_method = 0, and clusters = 2) represents the **average** data of 10 runs with those settings.  Each run randomly selected 15 of the 16 columns to cluster, then clustered each using fixed-member (quantiled) clustering into 2 values per column. You can see this provides great anonymity (all the way up to 10-anonymity), though our accuracy of the model is prob just out of range to meet the 95%CI threshhold

Notes on results values (2k - 10k).  As stated previously each row is an average result.  A 2k value represents the average ratio of rows that had **at least** (2-1) one other row with matching QI attribute values. A 10k value on the same row is generated from the same data and represents the ratio of rows that had at least (10-1) nine other rows with matching QI attributes.  Also, note that this algorithm only checks for matches within the clustered QI columns, the unclustered columns are not included.  This is why when the number of QIs is high, anonymity is much worse as it  becomes more difficult to create matching rows. 

Notes on accuracy and f-measure.  I say we disregard f-measure.  Accuracy is the accuracy of the average random forest model trained and executed on the settings shown by the grouping variables. The model is trained on the grouped QI data **and** the remaining ungrouped columns.  It is trained on 10% of the records (randomly chosen) and evaluated on 10% of the records (randomly chosen and dissimilar). Each random forest is composed of 20 trees who vote to determine the prediction. Bottom line is we see great durability in our model.  Only with extreme clustering - once with QIs very high (13-16), cluster_method = 0 and with low clusters - do we see it's performance start to drop off below 95% accuracy.  For reference the unclustered model has accuracy of 99.8%

**Results**:

1. Show which clustering technique provides better anonymity:  Fixed-member(cluster_method = 0) binning provides better anonymity than fixed-width (cluster_method = 1) binning.  This is true across the board.  The effect of each varies with the number of QIs (more QIs = worse anonymity) and the cluster sizes (more clusters = more granular data and worse anonymity), but the overall relationship remains (fixed-member better for this data-set)

2. Show how much better fixed-width binning is:  For instance,  3 QIs with fixed-member binning can handle up to 100 clusters and still provide better than .6 2-anonymity while fixed width-binning needs to be heavily clustered (10 clusters per QI) to provide >.6 2-anonymity.

3. Show how model results (accuracy) vary with number of QIs:  Model performs worse with more QIs.  This is due to the fact that more of the columns are clustered.

4. Show how model results vary with clustering method.  Not sure if one performs better in this regard, maybe method 1 gives better accuracy for given set of QIs and clusters (you can see a slight difference at high QIs when clusters is 2).

5. Show how model results vary with cluster size.  When there are few clusters (few = large) we see model results perform worse.  This is particularly visible when there at high QIS and true for both clustering methods. 

6. Putting it all together - show which combination of grouping variables provides best trade-off between anonymity and model performance.  If we set an anonymity cut-off at .9 and accuracy cut-off of .95 (per engels), what are the grouping variables that allows us the most QIs.  Answer:  With 10 QIs we are able to meet both thresholds.  This is using method 0, and very coarse clustering (clusters = 2).  Note the anonymity threshold is actually met for all values of k (2 - 10).  What about the other clustering technique? Cluster method 1 can meet both thresholds at QIs = 8, but only for 2-anonymity.  One has to go down to QIs = 5 to see method 1 provide 10-anonymity and meet both thresholds.
